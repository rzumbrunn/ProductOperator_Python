```python
import sys
import os
if '..' not in sys.path:
    sys.path.append('..')

from PO import *
PO.create(['I', 'S'])
from PO import *
import sympy as sp

tau = sp.symbols('tau', real=True, positive=True)

coupling_dir = [1,2,3]
rho_inits = [Ix,Iy, Iz]

for cd in coupling_dir:
    for rho0 in rho_inits:
        print("----------")
        print('cd = ', cd, 'rho0 = ', rho0)

        rho1 = rho0.jc(['IS'], [tau], cd)
        print('rho1 = ', rho1)

# homogeneous j coupling
def full_j(rho, value ,spins):
    for i in range(3):
        rho = rho.jc(spins, value, i)

    return rho

rho0 = Iz
rho = full_j(rho0, [tau], ['IS'])
print(rho)

print("-----")

rho0 = Iz
rho = rho0.jc(['IS'], [tau], -1)
print(rho)
```
